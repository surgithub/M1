```python
import numpy as np
import sympy as sy
from IPython.display import Latex, display
sy.init_printing() 
import matplotlib.pyplot as plt
display(Latex('Ex. Display the region: $1\leq r\leq 2$ and $0\leq \\theta\leq \\frac{\pi}{2}$.'))
display(Latex('Sol. It is the blue shaded region shown below.'))
plt.polar(np.deg2rad([0, 0]), [10,15],'r') # Not in the output but requited to use plt.fill
plt.fill_between(np.linspace(0, np.pi/2, 100), 1, 2, alpha=0.5,color='b')
plt.ylim(0,3.5) # r limits
plt.xticks((0,np.pi/2,np.pi,3*np.pi/2),\
('$0$','$\\frac{\pi}{2}$','$\pi$','$\\frac{3\pi}{2}$'), fontsize=20)
plt.yticks([1,2], fontsize=14)
plt.show()
```

```python
display(Latex('Ex. Display the region: $-3\leq r\leq 2$ and $\\theta=\\frac{\pi}{4}$.'))
display(Latex('Sol. It is the blue line segment shown below.'))
plt.polar(np.deg2rad([45, 45]), [0,2],'b') 
plt.polar(np.deg2rad([180+45, 180+45]), [0,3],'b')
plt.ylim(0,3.5) # r limits
plt.xticks((0,np.pi/2,np.pi/4,np.pi,5*np.pi/4,3*np.pi/2),\
('$0$','$\\frac{\pi}{2}$','$\\frac{\pi}{4}$','$\pi$','$\\frac{5\pi}{4}$','$\\frac{3\pi}{2}$'), fontsize=20)
plt.yticks([2,3], fontsize=14)
plt.show()
```

```python
display(Latex('Ex. Plot $r=1-\sin(\\theta/2)$.'))
display(Latex('Sol. It is the blue curve shown below.'))
theta = np.arange(0, 2*np.pi,0.01) # range of theta
plt.polar(theta, np.abs(np.sin(theta/2)),'b')
plt.polar(theta, np.abs(np.cos(theta/2)),'b')
plt.ylim(0,1.2) # r limits
plt.xticks((0,np.pi/2,np.pi,3*np.pi/2),\
('$0$','$\\frac{\pi}{2}$','$\pi$','$\\frac{3\pi}{2}$'), fontsize=20)
plt.yticks([1], fontsize=14)
plt.show()
```

```python
display(Latex('Ex. Find the area of the lemniscate $r^2=\cos2\\theta$.'))
display(Latex('Sol. It is the yellow shaded region shown below.'))
theta = np.arange(-np.pi/4, np.pi/4,0.0001) # range of theta
r=(np.abs(np.cos(2*theta)))**0.5
plt.polar(theta+(r<0)*np.pi, np.abs(r),'b')
plt.fill_between(theta, 0, r, alpha=0.5,color='y')
theta = np.arange(np.pi-np.pi/4, np.pi+np.pi/4,0.0001) # range of theta
r=(np.abs(np.cos(2*theta)))**0.5
plt.polar(theta+(r<0)*np.pi, np.abs(r),'b')
plt.fill_between(theta, 0, r, alpha=0.5,color='y')
plt.ylim(0,1.2) # r limits
plt.xticks((0,np.pi/4,np.pi/2,np.pi,3*np.pi/2,2*np.pi-np.pi/4),\
('$0$','$\\frac{\pi}{4}$','$\\frac{\pi}{2}$','$\pi$','$\\frac{3\pi}{2}$','$-\\frac{\pi}{4}$'), fontsize=20)
plt.yticks([1], fontsize=14)
plt.show()
t=sy.symbols('t')
display(Latex('This area reads as'))
display(sy.integrate(2*sy.cos(2*t),(t,0,sy.pi/4)))
```

```python
display(Latex('Ex. Find the area inside the circle $r=3 \cos\\theta$ but outside the cardioid $r=1+\cos\\theta$.'))
display(Latex('Sol. It is the yellow shaded region shown below.'))

theta = np.arange(0, 2*np.pi,0.0001) # range of theta
r1=np.abs(1+np.cos(theta))
r2=3*np.cos(theta)
plt.polar(theta+(r1<0)*np.pi, np.abs(r1),'b')
plt.polar(theta, r2,'r')
theta = np.arange(-np.pi/3, np.pi/3,0.0001) # range of theta
plt.fill_between(theta, 3*np.cos(theta),np.abs(1+np.cos(theta)), alpha=0.5,color='y')
plt.text(3*np.pi/2-0.8, 3.2, '$r=1+\cos\\theta$',fontsize=16,color='b')
plt.text(3*np.pi/2-0.6, 3.5, '$r=3\cos\\theta$',fontsize=16,color='r')
plt.ylim(0,3.5) # r limits
plt.xticks((0,np.pi/3,np.pi/2,np.pi,3*np.pi/2,2*np.pi-np.pi/3),\
('$0$','$\\frac{\pi}{3}$','$\\frac{\pi}{2}$','$\pi$','$\\frac{3\pi}{2}$','$-\\frac{\pi}{3}$'), fontsize=20)
plt.yticks([], fontsize=14)
plt.show()
t=sy.symbols('t')
display(Latex('Req. Area$=2\Big[\displaystyle\\frac{1}{2}\\int_{0}^{\pi/3}(3\cos\\theta)^2d\\theta-\\frac{1}{2}\int_{0}^{\pi/3}(1+\cos\\theta)^2d\\theta\Big]=$'))
display(sy.integrate((3*sy.cos(t))**2-(1+sy.cos(t))**2,(t,0,sy.pi/3)))
```
