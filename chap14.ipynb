```python
import numpy as np
import sympy as sy
from IPython.display import Latex, display, Math
sy.init_printing() 
import matplotlib.pyplot as plt
from sympy import Symbol, S
from sympy.calculus.util import continuous_domain
display(Latex('Ex. Find the domain of $f(x)=\sqrt{1-x^2}$, and also plot it.'))
display(Latex('Sol. Domain of the given function reads as'))
x = sy.symbols('x') 
f = sy.sqrt(1-x**2)
domain = continuous_domain(f, x, S.Reals)
display(domain)
display(Latex('The blue curve shown below is the plot of the given function.'))
import matplotlib.pyplot as plt
x = np.linspace(-1, 1, 100) 
#plt.figure(figsize=(4,3.5))
#plt.gca().set_aspect('1')
plt.plot(x,np.sqrt(1-x**2),'b')
#plt.plot(x,-np.sqrt(1-x**2),'b')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
display(Latex('Ex. Find $f_x$ and $f_y$ given that $f(x,y)=\\tan ^{-1}\\left(\\frac{x}{y}\\right)$.'))
display(Latex('Sol. We have'))
x,y, fx, fy=sy.symbols('x,y,f_x, f_y')
f=sy.atan(x/y)
display(sy.Eq(fx, sy.simplify(sy.diff(f,x)),evaluate=False))
display(sy.Eq(fy, sy.simplify(sy.diff(f,y)),evaluate=False))
```

```python
display(Latex('Ex. Show that $\displaystyle\\lim_{(x,y)\\rightarrow (0,0)}f(x,y)$, where \
$f(x,y)=\\frac{xy}{x^2+y^2}$, $(x,y)\\neq (0,0)$ does not exist.'))
display(Latex('Sol. Along the path $y=mx$, the given limit reads as'))
x,y,m,L=sy.symbols('x,y,m, L')
g=x*y/(x**2+y**2)
exp=sy.simplify(sy.limit(g,y,m*x))
#display(Latex(f"$lim_{(x,y)\rightarrow(0,0)}f(x,y)$={exp}"))
display(sy.Eq(L, exp, evaluate=False))

#display(fx)
if exp.has(x): display(Latex('It carries $x$, so the limit can not decided along the chosen path.'))
else: display(Latex('It depends on $m$. So the limit does not exist.'))
```

```python
i,j,k,x,y,z,t, L=sy.symbols('i,j,k,x,y,z, t, \\nabla{f}')
display(Latex('Ex. Find $\\nabla f$ at the point $(1,-2,-1)$, given that $f=3x^{2}y-y^{3}z^{2}$.'))
display(Latex('Sol. $\\nabla f$ at the point $(1,-2,-1)$ reads'))
f=3*x**2*y-y**3*z**2
gradf= sy.diff(f,x)*i+sy.diff(f,y)*j+sy.diff(f,z)*k
display(sy.Eq(L, gradf.subs([(x,1),(y,-2),(z,-1)]), evaluate=False))

#display(gradf.subs([(x,1),(y,-2),(z,-1)]))
```

```python
i,j,k,x,y,z,t=sy.symbols('i,j,k,x,y,z, t')
display(Latex('Ex. Find the directional derivative of  $f=x^{2}-y^{2}+2z^{2}$ at \
the point  $P(1,2,3)$ in the direction of line $PQ$, where $Q$ is \
the point (5,0,4). Also find the direction and magnitude of maximum directional derivative at $P$.'))
display(Latex('Sol. $\\nabla f$ at the point $P(1,2,3)$ reads'))
f=x**2-y**2+2*z**2
gradf= sy.diff(f,x)*i+sy.diff(f,y)*j+sy.diff(f,z)*k
gradfP=gradf.subs([(x,1),(y,2),(z,3)])
display(gradfP)
display(Latex('Unit vector $\\hat{n}$ at $P(1,2,3)$ in the direction of line $PQ$ is'))
ncap=(5-1)*i/sy.sqrt((5-1)**2+(0-2)**2+(4-3)**2)+(0-2)*j/sy.sqrt((5-1)**2+\
        (0-2)**2+(4-3)**2)+(4-3)*k/sy.sqrt((5-1)**2+(0-2)**2+(4-3)**2)
display(ncap)
display(Latex('So the directional derivative $\\nabla f.\\hat{n}$ at the point $P(1,2,3)$ reads'))
display(gradfP.coeff(i)*ncap.coeff(i)+gradfP.coeff(j)*ncap.coeff(j)+gradfP.coeff(k)*ncap.coeff(k))
display(Latex('Direction of the maximum directional derivative is $\\nabla f$, that is,'))
display(gradfP)
display(Latex('Magnitude of maximum directional derivative is $|\\nabla f|$, that is,'))
display(sy.sqrt(gradfP.coeff(i)**2+gradfP.coeff(j)**2+gradfP.coeff(k)**2))
```

```python
from sympy import series 
display(Latex('Ex. Find the Taylor series of $f(x)=\sin x$ about $x=\pi$, and upto $x^3$.'))
display(Latex('Sol. The required Taylor series reads as'))
x = sy.symbols("x")
display(sy.Eq(sy.sin(x), series(sy.sin(x), x, x0=sy.pi, n=7),evaluate=False))
```

```python
# MAXIMUM/MINIMUM VALUES OF ONE VARIABLE FUNCTION
display(Latex('Ex. Find the local minimum and maximum values of $f(x)=x^3-3x$.'))
display(Latex('Sol. The critical points are'))
f=x**3-3*x
sols=sy.solve([f.diff(x)], [x], dict=True)
g=f.diff(x,2)
h=f.diff(x,3)
#print(sols)
for i in range(0,len(sols)):
    print(sols[i])
for i in range(0,len(sols)):
    if g.subs(sols[i])>0:
        display (sols[i], 
               ' is a local minimum point and the corresponding local minimum value of f is', f.subs(sols[i]))
    elif g.subs(sols[i])<0:
        display (sols[i], 
               'is a local maximum point and the corresponding local maximum value of f is', f.subs(sols[i]))
    elif h.subs(sols[i])!=0: 
        display(sols[i], 'is point of inflection.')
    else: print('No conclusion based on second derivative test')
```

```python
# MAXIMUM/MINIMUM VALUES OF TWO VARIABLE FUNCTION
x,y=sy.symbols('x,y')
display(Latex('Ex. Find extremum values of $f(x)=3y^2-2y^3-3x^2+6xy$.'))
display(Latex('Sol. The critical points are'))
#f=3*y**2-2*y**3-3*x**2+6*x*y
f=10*x*y*sy.exp(-x**2-y**2)
r=f.diff(x,2);s=f.diff(x,y);t=f.diff(y,2);p=r*t-s**2
sols=sy.solve([f.diff(x), f.diff(y)], [x, y], dict=True)
display(sols)
#for i in range(0,len(sols)):
#    print(sols[i])
for i in range(0,len(sols)):
    if  r.subs(sols[i])>0 and p.subs(sols[i])>0:
        display(sols[i], 
               'is a local minimum point  and the corresponding local minimum value of f is', f.subs(sols[i]))
    elif r.subs(sols[i])<0 and p.subs(sols[i])>0:
        display(sols[i], "is a local maximum point and the corresponding local maximum value of f is", 
                f.subs(sols[i]))
    elif p.subs(sols[i])<0: 
        display(sols[i], 'is a saddle point and the corresponding value of f is', f.subs(sols[i]))
    else: print('No conclusion of the test. Further investigation is required')
```

```python
# LAGRANGE'S MULTIPLIER METHOD WITH ONE CONSTRAINT
x,y,z,k,a,b,c=sy.symbols('x,y,z,k,a,b,c')
f=x*y+2*y*z+2*z*x; g=x*y*z-32
#f=8*x*y*z; g=x**2/a**2+y**2/b**2+z**2/c**2-1

sols=sy.solve([f.diff(x)+k*g.diff(x),f.diff(y)+k*g.diff(y),f.diff(z)+k*g.diff(z),g], [x, y,z,k], dict=True)
display(sols)
for i in range(0,len(sols)):
    display(f.subs(sols[i]))
```

```python
# LAGRANGE'S MULTIPLIER METHOD WITH TWO CONSTRAINTS
x,y,z,k1,k2=sy.symbols('x,y,z,k1,k2')
f=x**2+y**2+z**2; g1=x**2+y**2-1; g2=x+y+z-1
sols=sy.solve([f.diff(x)+k1*g1.diff(x)+k2*g2.diff(x),f.diff(y)+
               k1*g1.diff(y)+k2*g2.diff(y),f.diff(z)+k1*g1.diff(z)+k2*g2.diff(z),g1,g2], [x, y,z,k1,k2], dict=True)
display(sols)
for i in range(0,len(sols)):
    display(f.subs(sols[i]))
```
