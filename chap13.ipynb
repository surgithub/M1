```python
import numpy as np
import sympy as sy
from IPython.display import Latex, display
sy.init_printing() 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

display(Latex('Ex. Plot the vector-valued function $\overrightarrow{r}(t)\
              =\cos t\;i+\sin t\;j+2\;k$, $t\in[0,2\pi]$.'))
display(Latex('Sol. It is a circle with centre at origin and radius 1 in the plane $z=2$ as shown below.'))
plt.gca(projection='3d')
t = np.linspace(0, 2*np.pi, 101)
plt.plot(np.cos(t), np.sin(t), 2, label\
    =r'Plot of the circle $\overrightarrow{r}(t)=\cos t\;i+\sin t\;j+2\;k$, $t\in[0,2]$')
plt.legend()
plt.show()
```

```python
display(Latex('Ex. Plot the vector-valued function $\overrightarrow{r}(t)\
              =\cos t\;i+\sin t\;j+t\;k$, $t\in[0,7\pi]$'))
display(Latex('Sol. It is a circular helix as shown below.'))
plt.gca(projection='3d')
t = np.linspace(0, 7*np.pi, 101)
plt.plot(np.cos(t), np.sin(t), t, label=\
    r'Plot of the circle $\overrightarrow{r}(t)=\cos t\;i+\sin t\;j+t\;k$, $t\in[0,7\pi]$')
plt.legend()
plt.show()
```

```python
i,j,k,x,y,z,t=sy.symbols('i,j,k,x,y,z, t')
display(Latex('Ex. Solve $\displaystyle\lim_{t\\rightarrow \pi/4}(\cos t\;i+\sin t\;j+t\;k)$.'))
display(Latex('Sol. It is given by'))
lim=sy.limit(sy.cos(t)*i+sy.sin(t)*j+t*k,t,sy.pi/4)
display(lim)
```

```python
i,j,k,x,y,z,t=sy.symbols('i,j,k,x,y,z, t')
display(Latex('Ex. A person on a hang glider is spiralling upward due rapidly\
rising air on a path $\overrightarrow{r}(t)=3\cos t\;\hat{i}+3\sin t\;\hat{j}+t^2\;\hat{k}$. \
Find the velocity vector, acceleration vector, gliders speed at time $t$, \
and the times when the gliders acceleration is orthogonal to its velocity.'))
display(Latex('Sol. Here velocity is'))
r=3*sy.cos(t)*i+3*sy.sin(t)*j+t**2*k
display(sy.diff(r,t))
display(Latex('Acceleration is'))
display(sy.diff(r,t,t))
display(Latex('Speed is'))
display(sy.simplify(sy.sqrt(((sy.diff(r,t).coeff(i))**2+(sy.diff(r,t).coeff(j))**2\
                             +(sy.diff(r,t).coeff(k))**2))))
display(Latex('Acceleration is orthogonal to velocity at'))
display(sy.solve(sy.diff(r,t).coeff(i)*sy.diff(r,t,t).coeff(i)+\
        +sy.diff(r,t).coeff(j)*sy.diff(r,t,t).coeff(j)+sy.diff(r,t).coeff(k)*sy.diff(r,t,t).coeff(k),t))
```

```python
i,j,k,x,y,z,t=sy.symbols('i,j,k,x,y,z, t')
display(Latex('Ex. Solve $\displaystyle \int_{0}^{\pi}(\cos t\;i+j-2t\;k)dt$'))
display(Latex('Sol. It is given by'))
r=sy.cos(t)*i+j-2*t*k
display(sy.integrate(r,(t,0,sy.pi)))
```

```python
i,j,k,x,y,z,t=sy.symbols('i,j,k,x,y,z, t')
display(Latex('Ex. A glider soaring upward along the helix \
$\overrightarrow{r}(t)=\cos t\;i+\sin t\;j+t\;k$. \
How far does the glider travel along its path from $t=0$ to $t=2\pi$.'))
display(Latex('Sol. It is given by'))
r=sy.cos(t)*i+sy.sin(t)*j+t*k
rm=sy.sqrt((sy.diff(r.coeff(i),t))**2+(sy.diff(r.coeff(j),t))**2+(sy.diff(r.coeff(k),t))**2)
display(sy.integrate(rm,(t,0,2*sy.pi)))
```

```python
i,j,k,a,b,t=sy.symbols('i,j,k,a,b,t')
display(Latex('Ex. Find the curvature and torsion of the helix \
$\overrightarrow{r}(t)=a\cos t\;i+a\sin t\;j+bt\;k$, $a,b\geq0$, $a^2+b^2\\neq0$.'))
display(Latex('Sol. $\kappa$ is given by'))
r=a*sy.cos(t)*i+a*sy.sin(t)*j+b*t*k
r1=sy.expand(sy.diff(r,t))
r2=sy.expand(sy.diff(r,t,t))
r3=sy.expand(sy.diff(r,t,t,t))
r1cr2m= sy.sqrt((r1.coeff(j)*r2.coeff(k)-r1.coeff(k)*r2.coeff(j))**2+\
(r1.coeff(k)*r2.coeff(i)-r1.coeff(i)*r2.coeff(k))**2+\
(r1.coeff(i)*r2.coeff(j)-r1.coeff(j)*r2.coeff(i))**2)
r1m=sy.sqrt((r1.coeff(i))**2+(r1.coeff(j))**2+(r1.coeff(k))**2)
display(sy.simplify(r1cr2m/r1m**3))

num_of_tau=r1.coeff(i)*(r2.coeff(j)*r3.coeff(k)-r2.coeff(k)*r3.coeff(j))+\
r1.coeff(j)*(r2.coeff(k)*r3.coeff(i)-r2.coeff(i)*r3.coeff(k))+\
r1.coeff(k)*(r2.coeff(i)*r3.coeff(j)-r2.coeff(j)*r3.coeff(i))

display(Latex('$\\tau$ is given by'))
display(sy.simplify(num_of_tau/r1cr2m**2))
```

```python
i,j,k,x,y,z,t=sy.symbols('i,j,k,x,y,z, t')
display(Latex('Ex. Find the tangential and normal components of the acceleration vector for the motion \
$\overrightarrow{r}(t)=(\sin t-t\cos t)\;i+(\cos t+t\sin t)\;j$. '))
display(Latex('Sol. Tangential component is'))
r=(sy.sin(t)-t*sy.cos(t))*i+(sy.cos(t)+t*sy.sin(t))*j
v=sy.diff(r,t)
a=sy.diff(r,t,t)
vm=sy.simplify(sy.sqrt(v.coeff(i)**2+v.coeff(j)**2+v.coeff(k)**2))
a_T=sy.diff(vm,t)
display(sy.powdenest(a_T,force=True))
am=sy.simplify(sy.sqrt(a.coeff(i)**2+a.coeff(j)**2+a.coeff(k)**2))
display(Latex('Normal component is'))
display(sy.powdenest(sy.sqrt(am**2-a_T**2),force=True))
```
